```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the GDP data
gdp_data = pd.read_csv('GDP.csv', parse_dates=['DATE'], index_col='DATE')
```

## 1. Study their time series properties: plot, descriptive statistics, ACF and PACF; do the same work for the first difference. Comment the results.

```python
# Plot the GDP time series
plt.figure(figsize=(8, 4))
plt.plot(gdp_data.index, gdp_data['GDP'], label='GDP')
plt.title('GDP Time Series')
plt.xlabel('Date')
plt.ylabel('GDP')
plt.legend()
plt.show()
```

```python
# Descriptive statistics
print("Descriptive Statistics for GDP:")
print(gdp_data.describe())
```

```python
# ACF and PACF plots for GDP
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
sm.graphics.tsa.plot_acf(gdp_data['GDP'], lags=50, ax=axes[0])
axes[0].set_title('ACF - GDP')
sm.graphics.tsa.plot_pacf(gdp_data['GDP'], lags=50, ax=axes[1])
axes[1].set_title('PACF - GDP')
plt.tight_layout()
plt.show()
```

```python
# First difference
gdp_diff = gdp_data['GDP'].diff().dropna()

# Plot the first difference
plt.figure(figsize=(10, 6))
plt.plot(gdp_diff.index, gdp_diff, label='First Difference of GDP')
plt.title('First Difference of GDP Time Series')
plt.xlabel('Date')
plt.ylabel('First Difference of GDP')
plt.legend()
plt.show()
```

```python
# Descriptive statistics for the first difference
print("Descriptive Statistics for First Difference of GDP:")
print(gdp_diff.describe())
```

```python
# ACF and PACF plots for the first difference
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
sm.graphics.tsa.plot_acf(gdp_diff, lags=50, ax=axes[0])
axes[0].set_title('ACF - First Difference of GDP')
sm.graphics.tsa.plot_pacf(gdp_diff, lags=50, ax=axes[1])
axes[1].set_title('PACF - First Difference of GDP')
plt.tight_layout()
plt.show()
```
